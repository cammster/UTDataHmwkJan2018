```python
#Import Statements,Twitter UserInformation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import tweepy
import time
from collections import OrderedDict
import seaborn as sns

# Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Twitter API Keys
consumer_key = "JovP5qnpxryY3o0G0dgoMuouc"
consumer_secret = "Cfvxna8OlA7IFnkRNGR4qJxtOzfhixJl0l5HXmJg0ocDgwaYwD"
access_token = "967431307176407046-oQRgCSqnwFHBJ4MuK3pq9LtkqflsWBN"
access_token_secret = "vQ313qjSD1ImvyMQU06zbmvMq9tiIxuA7wrlZfmSSIDRC"

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
#, parser=tweepy.parsers.JSONParser() use if not using Cursor method

#News Outlets and Colors

outlets=["@BBC","@CBS","@CNN","@foxnews","@nytimes"]
```

```python
# Loop through last 100 tweets for each news outlet,Perform Sentiment Analysis,Get Timestamp
timestamp=[]
text=[]
user=[]
compound=[]
pos=[]
neu=[]
neg=[]

for outlet in outlets:
   
    for item in tweepy.Cursor(api.user_timeline,id=outlet).items(100):
        tweetinfo = json.dumps(item._json,indent=5)
        tweet = json.loads(tweetinfo)

        text.append(tweet["text"])
        user.append(tweet["user"]["name"])
        timestamp.append(tweet["created_at"])
        compound.append(analyzer.polarity_scores(tweet["text"])["compound"])
        pos.append(analyzer.polarity_scores(tweet["text"])["pos"])
        neu.append(analyzer.polarity_scores(tweet["text"])["neu"])
        neg.append(analyzer.polarity_scores(tweet["text"])["neg"])
        
        
df=pd.DataFrame({"Text":text,"Timestamp":timestamp,
                             "User":user,"Compound":compound,"Pos":pos,"Neu":neu,"Neg":neg})
```

```python
df
newsoutlets=list(df["User"].unique())
newsoutlets
```

```python
#Plot1: Sentiment Analysis of Media Tweets(Date)
font={'family': 'arial',
        'color':  'black',
        'weight': 'bold',
        'size': 12}
colors=["lightblue","green","red","blue","yellow"]
counter1=0
counter2=0
counter3=0
counter4=0
counter5=0
plt.figure(figsize=(10,8))
sns.set_style("darkgrid")
for index,row in df.iterrows():
    
    for i in range(len(newsoutlets)):
        if (row["User"]==outlet(i)):
            counter1=counter1+1
            bbc=plt.scatter(counter1,row["Compound"],c=colors[i],edgecolor="black",alpha=0.5,s=70)
        if(row["User"]=="CBS"):
            counter2=counter2+1
            cbs=plt.scatter(counter2,row["Compound"],c=colors[1],edgecolor="black",alpha=0.5,s=70)
        if(row["User"]=="CNN"):
            counter3=counter3+1
            cnn=plt.scatter(counter3,row["Compound"],c="red",edgecolor="black",alpha=0.5,s=70)
        if(row["User"]=="Fox News"):
            counter4=counter4+1
            foxnews=plt.scatter(counter4,row["Compound"],c="blue",edgecolor="black",alpha=0.5,s=70)
        if(row["User"]=="The New York Times"):
            counter5=counter5+1
            nyt=plt.scatter(counter5,row["Compound"],c="yellow",edgecolor="black",alpha=0.5,s=70)
plt.gca().invert_xaxis()
plt.title("Sentiment Analysis of Media Tweets (3/5/2018)",fontdict=font)
plt.xlabel("Tweets Ago",fontdict=font)
plt.ylabel("Tweet Polarity",fontdict=font)
plt.legend(title="Media Sources",handles=[bbc,cbs,cnn,foxnews,nyt],fontsize=12,
           labels=newsoutlets,bbox_to_anchor=(1,0.5))
plt.show()
```
